```python
import json 

with open('./json_data/0.5 adv test acc.json', 'r') as file:
    # Load the JSON data
    adv_test_acc_data = json.load(file)
with open('./json_data/0.5 adv train acc.json', 'r') as file:
    # Load the JSON data
    adv_train_acc_data = json.load(file)
with open('./json_data/0.5 nat test acc.json', 'r') as file:
    # Load the JSON data
    nat_test_acc_data = json.load(file)
with open('./json_data/0.5 nat train acc.json', 'r') as file:
    # Load the JSON data
    nat_train_acc_data = json.load(file)
with open('./json_data/0.5 train loss.json', 'r') as file:
    # Load the JSON data
    train_loss_data = json.load(file)

model_name = "MNv3 small alpha=0.5"
print(f'{len(adv_train_acc_data)} {len(adv_test_acc_data)}')
```

```python
import matplotlib.pyplot as plt

# Sample data for four lines
nat_train_acc = nat_train_acc_data
nat_test_acc = nat_test_acc_data

# Extracting data for x and y axes for each line
nat_train_x = [entry[1] for entry in nat_train_acc]  # Epoch number
nat_train_y = [entry[2] for entry in nat_train_acc]  #   Value

nat_test_x = [entry[1] for entry in nat_test_acc] 
nat_test_y = [entry[2] for entry in nat_test_acc] 

# Plotting the lines
plt.plot(nat_train_x, nat_train_y, label='Natural Train Accuracy',)

plt.plot(nat_test_x, nat_test_y, label='Natural Test Accuracy')

# Adding labels and title
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.title(model_name + ' Natural Accuracy Curve')
plt.legend()  # Add legend if multiple lines are plotted

# Customizing grid style
plt.grid(True, linestyle='--')  # Dotted grid lines

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import json 

with open('./json_data/mnv3 train losss.json', 'r') as file:
    # Load the JSON data
    model_1_data = json.load(file)
with open('./json_data/pretrained train losss.json', 'r') as file:
    # Load the JSON data
    model_2_data = json.load(file)
with open('./json_data/frozen train losss.json', 'r') as file:
    # Load the JSON data
    model_3_data = json.load(file)
with open('./json_data/custom train losss.json', 'r') as file:
    # Load the JSON data
    model_4_data = json.load(file)

# Sample data for four lines
model_1 = model_1_data
model_2 = model_2_data
model_3 = model_3_data
model_4 = model_4_data

# Extracting data for x and y axes for each line

model_1_x = [entry[1] for entry in model_1] 
model_1_y = [entry[2] for entry in model_1] 

model_2_x = [entry[1] for entry in model_2] 
model_2_y = [entry[2] for entry in model_2] 

model_3_x = [entry[1] for entry in model_3] 
model_3_y = [entry[2] for entry in model_3] 

# Plotting the lines

plt.plot(model_1_x, model_1_y, label='MNv3_Small Train Loss')
plt.plot(model_2_x, model_2_y, label='Pre-trained MNv3_Small Train Loss')
plt.plot(model_3_x, model_3_y, label='Pre-trained MNv3_Small w/ Frozen Layers Train Loss')

# Adding labels and title
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.legend()  # Add legend if multiple lines are plotted
plt.xlim(0, 33)

# Customizing grid style
plt.grid(True, linestyle='--')  # Dotted grid lines

# Display the plot
plt.show()
```
